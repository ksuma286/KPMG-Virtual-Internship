```python
import pandas as pd
```

```python
df=pd.read_excel('1000 Marketing audience.xlsx')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.groupby('most_recent_buy_date')['total_spent'].sum().plot()
plt.xlabel('months of 2017')
plt.ylabel('Amount spent')
print('Distribution of total sales by top 1000 customers by months of 2017')
```

```python
sns.boxplot(data=df['total_spent'])
plt.title('Distribution of sales by top 1000 customers in 2017')
plt.ylabel('Amount spent')
```

```python
df['totsl_purchases'].hist()
plt.title('Distribution of total purchases by top 1000 customers in 2017')
plt.xlabel('Total Purchases')
plt.ylabel('Number of Customers')
```
