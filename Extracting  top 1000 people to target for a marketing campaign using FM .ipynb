```python
import pandas as pd
```

got 4  datasets CustomerDemographic,NewCustomerList,CustomerAddress,Transaction
Cleaned all the datasets and merged them into one single dataset KPMG
Using FM to find top 1000 customers to target for a marketing campaign
FM  means frequency of purchase or total purchases and total amount spent.

```python
df=pd.read_excel('KPMG.xlsx')
```

```python
df.tail(60)
```

```python
df.columns
```

```python
df.info()
```

```python
df['online_order'].head()
```

```python
df['online_order']=df['online_order'].astype('str')
```

```python
df['online_order']=df['online_order'].map({'0':'N','1':'Y'})
```

```python
df_a=df[df['deceased_indicator']=='N']
```

```python
df_a.head()
```

```python
df_a.shape
```

```python
df_a.duplicated().sum()
```

```python
df_total_sales_contribution=df_a.groupby('customer_id')['product_first_sold_price'].sum()
```

```python
df_total_sales_contribution.head()
```

```python
df_total_sales_contribution=df_total_sales_contribution.reset_index()
```

```python
df_total_sales_contribution.head()
```

```python
df_date_of_most_frequent_purchase=df_a.groupby('customer_id')['transaction_date'].max()
```

```python
df_date_of_most_frequent_purchase=df_date_of_most_frequent_purchase.reset_index()
```

```python
df_date_of_most_frequent_purchase.head()
```

```python
df_total_purchases_2017=df_a.groupby('customer_id')['product_id'].count()
```

```python
df_total_purchases_2017=df_total_purchases_2017.reset_index()
```

```python
df_total_purchases_2017.head()
```

```python
df_total_purchases_past_three_years=df_a[['customer_id','past_3_years_bike_related_purchases']]
```

```python
df_total_purchases_past_three_years.head()
```

```python
df_total_purchases_past_three_years.drop_duplicates(inplace=True)
```

```python
df_total_purchases_past_three_years.head()
```

```python
dff=pd.merge(df_total_sales_contribution,df_date_of_most_frequent_purchase,on='customer_id')
```

```python
dff=pd.merge(dff,df_total_purchases_2017,on='customer_id')
```

```python
dff=pd.merge(dff,df_total_purchases_past_three_years,on='customer_id')
```

```python
dff.head()
```

```python
dff.rename(columns={'product_first_sold_price':'total_spent','transaction_date':'most_recent_buy_date','product_id':'totsl_purchases'},inplace=True)
```

```python
dff.head()
```

```python
dff=dff.sort_values(by=['total_spent','totsl_purchases','past_3_years_bike_related_purchases'],ascending=[False,False,False])
```

```python
#sorting by multiple values
```

```python
dff.head()
```

```python
dff_top_1000=dff.iloc[0:1000,:]
```

```python
dff_top_1000.tail()
```
